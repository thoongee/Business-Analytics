## animal shelter

count '보호소 수', '폐사안락사수' for each dong and calculate '보호소 밀도', '폐사안락사비율'

```python
ani = pd.read_excel('animal_shelter.xlsx')
ani
```

```python
ani.drop(["동별(1)","구분별(1)","구분별(2)"],axis=1,inplace=True)
ani
```

```python
# fill all Nan values into 0
ani = ani.fillna(0)

animal = ani.fillna(0)
animal
```

```python
# change each column's 0 values with its above letter (ex.00구, 00동) 
# initial letter =  '소계', becuase first row of each columns has '소계' values

dong = "소계"
for j in range(0,ani.shape[1]-1):
    for i in range(0,len(ani)):
        if(ani.iloc[i,j])== 0:
            ani.iloc[i,j] = dong
        else:
            dong = ani.iloc[i,j]
```

```python
ani
```

```python
# initialize index to see comfortably
ani = ani.reset_index(drop=True)
ani
```

```python
# change column names
ani.columns=['구','서울시 동','종류','폐사안락사수']
```

```python
ani
```

```python
# only leave rows which have '종류' == '폐사안락사'
list = []
for i in range(0,len(ani)):
    if(ani.iloc[i,2])=='폐사안락사':
        list.append(i)

ani = ani.iloc[list,:]
ani
```

```python
# remove '종류' column
ani.drop(["종류"],axis=1,inplace=True)
```

```python
ani = ani.reset_index(drop=True)
```

```python
ani
```

```python
# remove '구'=='소계' rows
animal = ani.drop(index=[0,1,2],axis=0)
animal
```

```python
# change '...' values to 0 
animal = animal.replace('...',0)
```

```python
animal = animal.astype({'폐사안락사수': 'int'})
```

```python
animal.dtypes
```

```python
animal
```

```python
delete = animal['서울시 동'].isin(['소계'])
animal = animal[~delete]
animal = animal.reset_index(drop=True)
animal
```

```python
# Combine the number of animal deaths in groups of "Dong"
animal = animal.groupby("서울시 동")['폐사안락사수'].sum().to_frame()
```

```python
animal
```

```python
# make new column
animal["보호소 수"]=0
animal
```

```python
# 보호소 수 : if '폐사안락사수' != 0 than '보호소 수' = 1, else 0
for i in range(0,len(animal)):
    if(animal.iloc[i,0])!=0:
      animal.iloc[i,1]=1

animal
```

```python
animal.to_excel('animal_shelter_result.xlsx')
```
